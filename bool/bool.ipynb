## Truth Table Generator


You can use `truth-table-generator` to create a truth table. First, intall the package.

And import the package.

```python
import ttg
```

A truth table has one column for each input variable (for example, p and q). If the input has only one list of strings, each string is considered an input variable:

```python
print(ttg.Truths(['p', 'q', 'r']))
```

If you prefer the words True and False instead of numbers 0 and 1, there is a third parameter, boolean type, ints that can be set to False:

```python
print(ttg.Truths(['p', 'q', 'r'], ints=False))
```

A second list of strings can be passed with propositional expressions created with logical operators.

```python
print(ttg.Truths(['p', 'q'], ['not p', 'p and q', 'p or q'], ints=False))
```

It supports the following operators:

- negation: `not`, `-`, `~`
- disjunction: `or`
- conjunction: `and`
- conditional (material implication): `=>`, `implies`
- biconditional: '='
- `nor`
- exclusive disjunction: `xor`, `!=`
- `nand`

```python
print(ttg.Truths(['p', 'q'], ['p => q', 'p = q', 'p or q', 'p nor q', 'p xor q', 'p and q', 'p nand q'], ints=False))
```

```python
print(ttg.Truths(['A', 'B'], ['A and B', 'A'], ints=False))
```

```python
print(ttg.Truths(['p', 'q'], ['(p and p) => p']))
```
